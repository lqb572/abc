```python
driver.find_element_by_xpath(r'//*[@id="detail"]/div[1]/ul/li[5]').click()#点击累计评价
```

```python
reobj = re.compile(r'<div class="comment-item"([\s\S]*?)</div></div>')#使用正则提取网页中有<div class="comment-item"([\s\S]*?)</div></div>标签的内容
```

```python
reobj1 = re.compile(r'p class="comment-con">([\s\S]*?)</p>')#二次使用正则表达式提取p class="comment-con">([\s\S]*?)</p>的内容，最终提取到评论
```

```python
comment = []
count = 1#建立一个comment
```

```python
def next_page(driver):
    # driver.find_element_by_link_text(u"下一页").click()
    parent = driver.find_element_by_css_selector('.ui-page')
    try:
        parent.find_element_by_link_text(u"下一页").click()
    except:
        time.sleep(2)
        page_num = parent.find_element_by_link_text(u"下一页").get_attribute('rel') 
        # ActionChains(driver).send_keys(Keys.SPACE).perform()
        # time.sleep(2)
        parent.find_element_by_link_text(page_num).click()
        # driver.find_element_by_link_text(u"下一页").click()
```

```python
while True:
    html = driver.page_source   #获取源代码
    result = reobj.finditer(html)   #用一次正则查找源码中的信息，将其保存到result中
    
    for i in result:   #遍历一次正则找到的信息
        z = reobj1.search(i.group(1))  #用二次正则匹配评论数据，将一条评论数据存入z变量中
        try:
            comment.append(z.group(1)) #如果匹配成功，将z加入comment中
            print(str(count) + ".\t" + z.group(1))
            count += 1
        except:
            pass
    retry = 0
    
    try:   #尝试翻页两次，如果不成功，跳出循环，程序结束
        time.sleep(2)
        next_page(driver)
    except:
        try:
            next_page(driver)
        except:
            break
        
        
##-----------------点击下一页的另一种思路，找到点击下一页那一组的所有元素，遍历后点击最后一个
#    for p in Page_turning.find_elements_by_tag_name("a"):
#        ActionChains(d).move_to_element(p).perform()
#    retry = 0
#    for j in range(0,5):
#        try:
#            time.sleep(2)
#            p.click()
#        except:
#            retry += 1
#    if retry > 4:
#        print(retry)
#        break                
```
