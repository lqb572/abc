```python
reobj = re.compile(r'target="_blank" seed="consumeTitle-linkT[\d\D]*?" smartracker="on">([\d\D]*?)</a>')
```

```python
def next_page(driver):
    driver.find_element_by_css_selector(".page-next.page-trigger").click()
```

```python
comment = []     #建立一个comment
count = 1
```

```python
e=1
```

```python
next_page(driver)
```

```python
driver.find_element_by_css_selector(".page-next.page-trigger")
```

```python
while True:
    html = driver.page_source     #获取源代码
    result = reobj.finditer(html)     #用一次正则查找源码中的信息，将其保存到result中
    if not result:
        print("找不到评论")    #如果没有评论，输出“找不到评论”
    for i in result:
        comment.append(i.group(1))       #如果匹配成功，将i加入comment中
        print(str(count) + ".\t" + i.group(1))
        count += 1
```

```python
try:         #尝试翻页两次，如果不成功，跳出循环，程序结束
        next_page(driver)
    except:
        try:
            next_page(driver)
        except:
            break
```
