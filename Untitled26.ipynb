```python
from selenium.webdriver.common.keys import Keys
import subprocess
import codecs
from selenium import webdriver
import re
import time
from selenium.webdriver.chrome.options import Options
url="https://item.jd.com/34640640658.html"
d = webdriver.Chrome() 
d.get(url)
time.sleep(3)
d.find_element_by_xpath(r'//*[@id="detail"]/div[1]/ul/li[5]').click()
reobj = re.compile(r'<p class="comment-con">([\s\S]*?)</p>')
comment = []
count = 1
def next_page(d):
    try:
        d.find_element_by_css_selector('.ui-page').find_element_by_link_text(u"下一页").click()
        time.sleep(3)
    except:
        pass
while True:
    html = d.page_source
    result = reobj.finditer(html)
    for i in result:
        comment.append(i.group(1))
        print(str(count) + ".\t" + i.group(1))
        count += 1  
    next_page(d)
    #sreach_window=d.current_window_handle
    if(count>=50):
        break
```
