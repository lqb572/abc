```python
from selenium import webdriver
import re
driver=webdriver.Chrome()
list = []
for i in range(1,10):
    url = 'https://www.zhipin.com/c101210100/?query=web前端&page='+str(i)
    print(url)
    driver.get(url)
#     print(driver.page_source)
    html = driver.page_source
    reobj = re.compile(r"""<li>
                                    <div class="job-primary">
                                        <div class="info-primary">
                                            <h3 class="name">[\s\S]*?</li>""")
    count = 0
    print("================================")
    reobj = re.compile(r"<li>[\s\S]*?</li>")
    for match in reobj.finditer(html):
        li = match.group(0)
        if li.find("job-primary") > -1:
            job_title = ""
            company = ""
            match2 = re.search(r'<div class="job-title">([\s\S]*?)</div>', li)
            if match2:
                job_title = match2.group(1)
                print(job_title)
                
            match3 = re.search(r'<div class="company-text">[\s\S]*?"_blank">([\s\S]*?)</a>', li)
            if match3:
                company = match3.group(1)
                print(company)
            match4 = re.search(r'<span class="red">([\s\S]*?)</span>', li)
            if match4:
                red = match4.group(1)
                print(red)
            match5 = re.search(r'<p>([\s\S]*?)<em class="vline">([\s\S]*?)</p>',li)
            if match5:
                vline = match5.group(1)
                print(vline)
            match6 = re.search(r'<em class="vline"></em>([\s\S]*?)</p>',li)
            if match6:
                vline = match6.group(1)
                print(vline)
            list.append((job_title,company,red,vline,vline))
print(list)

```
